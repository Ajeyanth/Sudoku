You must include a function called `sudoku_solver(sudoku)` which takes one sudoku puzzle (a 9x9 NumPy array) as input, and returns the solved sudoku as another 9x9 NumPy array. This is the function which will be tested. 

```python
import numpy as np
import itertools
from collections import defaultdict


class SudokuSolver:
    def __init__(self):
        self.lcv_cache = {}
        self.all_units = self.get_sudoku_units()

    def work_out_lcv(self, sudoku, candidates, row, col, num):
        # Check if the result is cached
        if (row, col, num) in self.lcv_cache:
            return self.lcv_cache[(row, col, num)]

        constraint_count = 0
        # Check same row and column
        for i in range(9):
            if i != col and num in candidates[row][i]:
                constraint_count += 1
            if i != row and num in candidates[i][col]:
                constraint_count += 1

        # Check same box
        box_row_start, box_col_start = 3 * (row // 3), 3 * (col // 3)
        for i in range(box_row_start, box_row_start + 3):
            for j in range(box_col_start, box_col_start + 3):
                if (i != row or j != col) and num in candidates[i][j]:
                    constraint_count += 1

        # Cache the result
        self.lcv_cache[(row, col, num)] = constraint_count
        return constraint_count

    def clear_cache(self):
        self.lcv_cache.clear()

    def get_sudoku_units(self):
        units = []

        # Rows
        for i in range(9):
            units.append([(i, j) for j in range(9)])

        # Columns
        for j in range(9):
            units.append([(i, j) for i in range(9)])

        # 3x3 Boxes
        for box_row in range(0, 9, 3):
            for box_col in range(0, 9, 3):
                units.append([(box_row + i, box_col + j) for i in range(3) for j in range(3)])

        return units


lcv_solver = SudokuSolver()


def valid_move_check(sudoku, row, col, num):
    box_row = row - row % 3
    box_col = col - col % 3
    if num in sudoku[row] or num in sudoku[:, col] or num in sudoku[box_row:box_row + 3, box_col:box_col + 3].flatten():
        return False
    return True


def valid_puzzle_check(sudoku):
    """
    Validates if the given sudoku puzzle is correct according to Sudoku rules.

    This function checks every row, column, and 3x3 box to ensure that there are no duplicate
    numbers. Only non-zero elements are considered since zeros represent empty cells.

    """
    for i in range(9):
        row = sudoku[i, :]
        col = sudoku[:, i]
        box = sudoku[3 * (i // 3):3 * (i // 3) + 3, 3 * (i % 3):3 * (i % 3) + 3].flatten()

        if np.unique(row[row > 0]).size != row[row > 0].size:
            return False
        if np.unique(col[col > 0]).size != col[col > 0].size:
            return False
        if np.unique(box[box > 0]).size != box[box > 0].size:
            return False
    return True


def find_initial_candidates(sudoku):
    """
    Initialises the candidates for each cell in the Sudoku puzzle.

    Each cell gets a set of possible numbers (1-9) if it's empty (denoted by 0), or an empty set if it already
    has a number. This is used to track the allowed values for each cell as the puzzle is being solved.

    """
    candidates = [[set(range(1, 10)) if sudoku[i][j] == 0 else set() for j in range(9)] for i in range(9)]
    for i in range(9):
        for j in range(9):
            if sudoku[i][j] != 0:
                # Update the candidates immediately based on the initial numbers on the board.
                update_candidates(candidates, i, j, sudoku[i][j], None)
    return candidates


def update_candidates(candidates, row, col, num, history=None):
    """
    Updates the candidates for cells affected by placing a number in the given cell.

    This function removes the number from the candidates of all cells in the same row, column,
    and 3x3 box. It keeps track of changes for possible backtracking.

    """
    changes = []  # To keep track of the changes for backtracking.

    # Update candidates in the same row and column.
    for i in range(9):
        if num in candidates[row][i] and i != col:
            changes.append(((row, i), candidates[row][i].copy()))
            candidates[row][i].discard(num)

        if num in candidates[i][col] and i != row:
            changes.append(((i, col), candidates[i][col].copy()))
            candidates[i][col].discard(num)

    # Update candidates in the same 3x3 box.
    box_row_start, box_col_start = 3 * (row // 3), 3 * (col // 3)
    for i in range(box_row_start, box_row_start + 3):
        for j in range(box_col_start, box_col_start + 3):
            if num in candidates[i][j] and (i, j) != (row, col):
                changes.append(((i, j), candidates[i][j].copy()))
                candidates[i][j].discard(num)

    # Record changes if history is being kept.
    if history is not None:
        history.append(changes)


def reset_candidates(candidates, history):
    """
    Resets the candidates for cells to a previous state.

    This function is used for backtracking, where it reverts the candidates to their state before
    the last attempted number placement.

    """
    last_changes = history.pop()  # Get the last set of changes to undo.
    for cell, prev_candidates in last_changes:
        candidates[cell[0]][cell[1]] = prev_candidates


def discard_row_outside_box(sudoku, candidates, row, box_col, num):
    changed = False
    for col in range(9):
        if box_col <= col < box_col + 3:
            continue
        if num in candidates[row][col]:
            candidates[row][col].discard(num)
            changed = True
    return changed


def discard_column_outside_box(sudoku, candidates, col, box_row, num):
    changed = False
    for row in range(9):
        if box_row <= row < box_row + 3:
            continue
        if num in candidates[row][col]:
            candidates[row][col].discard(num)
            changed = True
    return changed


def pointing_pairs(sudoku, candidates):
    made_progress = False

    for box_row in range(0, 9, 3):
        for box_col in range(0, 9, 3):
            for num in range(1, 10):
                cells = [(r, c) for r in range(box_row, box_row + 3) for c in range(box_col, box_col + 3) if
                         num in candidates[r][c]]
                if not cells:
                    continue

                rows, cols = zip(*cells)
                unique_rows, unique_cols = set(rows), set(cols)

                if len(unique_rows) == 1:
                    made_progress |= discard_row_outside_box(sudoku, candidates, unique_rows.pop(), box_col, num)

                if len(unique_cols) == 1:
                    made_progress |= discard_column_outside_box(sudoku, candidates, unique_cols.pop(), box_row, num)

    return made_progress


def naked_pairs(candidates):
    made_progress = False

    for unit in lcv_solver.all_units:
        # Check each combination of two cells within the unit
        for cell1, cell2 in itertools.combinations(unit, 2):
            if candidates[cell1[0]][cell1[1]] == candidates[cell2[0]][cell2[1]] and len(
                    candidates[cell1[0]][cell1[1]]) == 2:
                # Found a naked pair
                for cell in unit:
                    if cell != cell1 and cell != cell2:
                        # Remove these candidates from other cells in the unit
                        old_candidates = candidates[cell[0]][cell[1]].copy()
                        candidates[cell[0]][cell[1]] -= candidates[cell1[0]][cell1[1]]
                        if old_candidates != candidates[cell[0]][cell[1]]:
                            made_progress = True

    return made_progress


def hidden_pairs(candidates):
    made_progress = False

    for unit in lcv_solver.all_units:
        # Track the positions of each candidate number
        candidate_positions = defaultdict(list)
        for cell in unit:
            row, col = cell
            for num in candidates[row][col]:
                candidate_positions[num].append(cell)

        # Look for hidden pairs among the candidates
        for nums in itertools.combinations(range(1, 10), 2):
            common_cells = set(candidate_positions[nums[0]]) & set(candidate_positions[nums[1]])
            if len(common_cells) == 2:
                # Potential hidden pair found
                # Check if these are the only two cells where both nums appear
                if all(len(candidate_positions[num]) == 2 for num in nums):
                    for cell in common_cells:
                        row, col = cell
                        if candidates[row][col] != set(nums):
                            candidates[row][col] = set(nums)
                            made_progress = True

    return made_progress


def naked_singles(sudoku, candidates):
    made_progress = False
    single_candidate_cells = [(i, j) for i in range(9) for j in range(9) if len(candidates[i][j]) == 1]

    while single_candidate_cells:
        row, col = single_candidate_cells.pop()
        if sudoku[row][col] == 0 and len(candidates[row][col]) == 1:
            value = candidates[row][col].pop()
            sudoku[row][col] = value
            update_candidates_and_track_changes(candidates, row, col, value, single_candidate_cells)
            made_progress = True

    return made_progress


def update_candidates_and_track_changes(candidates, row, col, num, single_candidate_cells):
    """
    Updates the candidates for each cell after a number placement and tracks cells that
    become naked singles as a result.

    """
    # Update row candidates and add single candidates to tracking list
    for i in range(9):
        if i != col and num in candidates[row][i]:
            candidates[row][i].discard(num)
            if len(candidates[row][i]) == 1:
                single_candidate_cells.append((row, i))

    # Update column candidates and add single candidates to tracking list
    for i in range(9):
        if i != row and num in candidates[i][col]:
            candidates[i][col].discard(num)
            if len(candidates[i][col]) == 1:
                single_candidate_cells.append((i, col))

    # Update box candidates and add single candidates to tracking list
    box_row_start, box_col_start = 3 * (row // 3), 3 * (col // 3)
    for i in range(box_row_start, box_row_start + 3):
        for j in range(box_col_start, box_col_start + 3):
            if (i, j) != (row, col) and num in candidates[i][j]:
                candidates[i][j].discard(num)
                if len(candidates[i][j]) == 1:
                    single_candidate_cells.append((i, j))


def hidden_singles(sudoku, candidates):
    """
    Applies the hidden singles technique to the puzzle.

    This technique finds cells where a candidate number is the only possibility
    within a unit (row, column, or box) and fills in the number.

    """
    made_progress = False
    for unit in lcv_solver.all_units:
        for num in range(1, 10):
            cells = [cell for cell in unit if num in candidates[cell[0]][cell[1]]]
            if len(cells) == 1:
                row, col = cells[0]
                sudoku[row][col] = num
                candidates[row][col].clear()
                update_candidates(candidates, row, col, num)
                made_progress = True
    return made_progress


def sudoku_solver(sudoku):
    """
    Solves a Sudoku puzzle using constraint propagation and backtracking search.

    """
    candidates = find_initial_candidates(sudoku)
    history = []
    if not valid_puzzle_check(sudoku):
        return np.full((9, 9), -1)

    sudoku = initial_constraint_propagation(sudoku, candidates)
    # If puzzle is still not solved, use DFS backtracking with MRV
    if not check_is_sudoku_solved(sudoku):
        sudoku = finish_sudoku(sudoku, candidates, history)
    return sudoku


def initial_constraint_propagation(sudoku, candidates):
    # Apply naked and hidden singles first
    while True:
        made_progress = False

        made_progress = naked_singles(sudoku, candidates)
        made_progress = hidden_singles(sudoku, candidates)

        naked_pairs(candidates)

        if pointing_pairs(sudoku, candidates):
            made_progress = True

        hidden_pairs(candidates)

        if not made_progress:
            break
    return sudoku


def check_is_sudoku_solved(sudoku):
    """
    Checks if a Sudoku puzzle is solved.

    """
    return all(sudoku[i][j] != 0 for i in range(9) for j in range(9))


def find_best_cell_mrv(sudoku, candidates):
    """
    Finds the next empty cell with the minimum remaining possible values (MRV heuristic).

    """
    min_remaining_values = float('inf')
    selected_cell = None

    # Loop through each cell to find the one with the fewest candidates
    for i in range(9):
        for j in range(9):
            if sudoku[i][j] == 0:  # If the cell is empty
                num_candidates = len(candidates[i][j])
                if num_candidates < min_remaining_values:
                    min_remaining_values = num_candidates
                    selected_cell = (i, j)

    return selected_cell


def finish_sudoku(sudoku, candidates, history):
    """
    Solves the Sudoku puzzle using backtracking search with MRV and LCV heuristics.

    """
    # Find the next cell to fill based on the MRV heuristic
    mrv_cell = find_best_cell_mrv(sudoku, candidates)
    if mrv_cell is None:  # If there's no empty cell left, the puzzle is solved
        return np.array(sudoku) if check_is_sudoku_solved(sudoku) else np.full((9, 9), -1)

    row, col = mrv_cell
    # Sort candidates based on the LCV heuristic
    sorted_candidates = sorted(
        candidates[row][col],
        key=lambda num: lcv_solver.work_out_lcv(sudoku, candidates, row, col, num)
    )

    for num in sorted_candidates:
        # Try placing each candidate and use recursion.
        if valid_move_check(sudoku, row, col, num):
            sudoku[row][col] = num
            update_candidates(candidates, row, col, num, history)
            result = finish_sudoku(sudoku, candidates, history)
            if result is not None and not np.array_equal(result, np.full((9, 9), -1)):
                return result  # If a valid solution is found, return it

            # If the choice leads to no solution, backtrack
            sudoku[row][col] = 0
            reset_candidates(candidates, history)

    # If no candidates work, return invalid
    return np.full((9, 9), -1)



```
